```python
import numpy as np

# New true & predicted values
actual = np.array([5.2, 3.8, 7.1, 4.5, 6.6])
pred = np.array([4.9, 4.1, 6.9, 5.0, 6.3])

print("Actual Values:", actual)
print("Predicted Values:", pred)
```

```python
# 1. Mean Squared Error (MSE)
mse_loss = np.mean((actual - pred)**2)
print("Mean Squared Error:", mse_loss)
```

```python
# 2. Mean Absolute Error (MAE)
mae_loss = np.mean(np.abs(actual - pred))
print("Mean Absolute Error:", mae_loss)
```

```python
# 3. Root Mean Squared Error (RMSE)
rmse_loss = np.sqrt(mse_loss)
print("RMSE:", rmse_loss)
```

```python
# 4. Binary Cross Entropy (Example values converted to binary format)
y_true_bin = np.array([1,0,1,0,1])
y_pred_bin = np.array([0.92,0.25,0.88,0.30,0.70])

bce_loss = -np.mean(y_true_bin*np.log(y_pred_bin) + (1-y_true_bin)*np.log(1-y_pred_bin))
print("Binary Cross Entropy Loss:", bce_loss)
```

```python
# 5. Categorical Cross Entropy (Example multi-class)
y_true_cat = np.array([[1,0,0],[0,1,0],[0,0,1]])
y_pred_cat = np.array([[0.85,0.10,0.05],
                       [0.20,0.70,0.10],
                       [0.05,0.10,0.85]])

cce_loss = -np.mean(np.sum(y_true_cat * np.log(y_pred_cat), axis=1))
print("Categorical Cross Entropy Loss:", cce_loss)
```
